# PEMFC Polarization Curve — Steady-State Solver

Same cell configuration as `simulate_polarization_curve_channel.ipynb` but the
galvanostatic sweep uses `BaseModel.solve_steady_state` (a root finder on
`rates_of_change = 0`) instead of time-integration.

**Advantages over IVP sweep:** no need to integrate 500 s per current step — each
operating point is found directly by the nonlinear solver, warm-started from the
previous solution.  **Limitations:** the solver may not converge near the limit
current where the Jacobian becomes ill-conditioned.


```python
import numpy as np
import cantera as ct
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

import marapendi.dynamic as mrpd
```

## 1 — Cell assembly

All parameters follow Vetter & Schumacher (2019) Tables 3 and 4.

```python
# ── Kinetics ──────────────────────────────────────────────────────────────
# ORR: Neyerlin et al. (2006); Eq. 18 of Vetter & Schumacher (2019).
# activation_energy in J/kmol (Cantera convention): 67 kJ/mol = 67e6 J/kmol.
orr_kinetics = mrpd.ElectrochemicalReaction(
    reference_exchange_current_density=2.45e-4,   # A/m2_Pt
    activation_energy=67e6,                        # J/kmol = 67 kJ/mol
    reaction_order=0.54,
    reference_activity=1.e5,                       # Pa (1 atm reference)
    reference_temperature=353.15,                  # K  (80 C)
    number_of_electrons=1,
    charge_transfer_coeff=1,
)


# ── GDLs — Toray TGP-H-060 (Table 4) ─────────────────────────────────────
gdl_an = mrpd.PorousLayer(
    thickness=160e-6, eps_p=0.76, bulk_density=440.,
    bulk_specific_heat_capacity=710., bulk_thermal_conductivity=1.6,
    K_abs=6.15e-12, theta_contact=130., tort=1.6**2
)
gdl_ca = mrpd.PorousLayer(
    thickness=160e-6, eps_p=0.76, bulk_density=440.,
    bulk_specific_heat_capacity=710., bulk_thermal_conductivity=1.6,
    K_abs=6.15e-12, theta_contact=130., tort=1.6**2
)

# ── Nafion NR-211 ionomer (Section 3.3) ──────────────────────────────────
# EW = 1100 g/mol, rho_dry = 2000 kg/m3.
# Sorption isotherm: Springer et al. (1991) Eq. 24.
# Desorption coefficient a_d = 1.42e-2 cm/s; absorption a_a = 3.53e-3 cm/s
# (Ge et al. 2005, cited as Eq. 25 in Vetter & Schumacher).
nafion = mrpd.PFSAIonomer(
    rho_dry_ion=1.97e3, EW_ion=1020,
    # Vetter & Schumacher (2019) Eq. 22, fit to Mittelsteadt & Staser (2011)
    darken_num_ion=np.array([0., 67.74, -32.03, 3.842]),          # ascending degree: c0 + c1*lmbd + c2*lmbd^2 + c3*lmbd^3
    darken_den_ion=np.array([103.37, -33.013, -2.115, 1.0]),       # ascending degree
    sorption_coeffs_ion=np.array([0.043, 17.81, -39.85, 36.0]),
    lmbd_liq_ref_ion=22,
    D_lmbd_ref_ion=1.0e-10,        # 10^-6 cm^2/s = 10^-10 m^2/s (Vetter Eq. 22)
    k_des_ref_ion=1.42e-4,   # a_d = 1.42e-2 cm/s -> 1.42e-4 m/s
    E_act_ion=20e6,           # J/kmol = 20 kJ/mol
    E_act_cond_ion=15e6,
    sigma_ref_ion=116.,
    f_v_perc_ion=0.06, n_sigma_ion=1.5,
    T_ref_sigma_ion=353.15, T_ref_D_ion=353.15, T_ref_des_ion=353.15,
)


# ── Catalyst layers ────────────────────────────────────────────────────────
# Cathode: Pt loading 3x anode -> a_ca = 3e11 cm2_Pt/m3  (Table 4)
ca_cl = mrpd.PtCCatalystLayer(
    thickness=10e-6,
    bulk_density=1000., bulk_specific_heat_capacity=710.,
    bulk_thermal_conductivity=0.27,
    L_Pt=0.4e-2, wt_Pt=0.416, ic_ratio=1.04, ecsa=75e3, 
    tort=1.6**2, ionomer=nafion,
    r_C=1e-10, K_abs=1e-13, theta_contact=95,
    reaction=orr_kinetics
)
# Anode: 1/3 Pt loading -> a_an = 1e11 cm2_Pt/m3
an_cl = mrpd.PtCCatalystLayer(
    thickness=10e-6,
    bulk_density=1000., bulk_specific_heat_capacity=710.,
    bulk_thermal_conductivity=0.27,
    L_Pt=0.4e-2 / 3, wt_Pt=0.192, ic_ratio=1.07, ecsa=75e3,
    tort=1.6**2, ionomer=nafion,
    r_C=1e-10, K_abs=1e-13, theta_contact=95
)

# ── Cell assembly ──────────────────────────────────────────────────────────
cell = mrpd.Cell(
    area=50e-4,
    electrical_resistance=80e-6/1250. + 5e-6/350.,
    thermal_resistance=0,
    ca=mrpd.CellSide(
        cl=ca_cl, gdl=gdl_ca,
        ch=mrpd.FlowChannel(height=1e-3, bulk_thermal_conductivity=100., width=1e-3, n_parallel=100, length=50e-4/(2e-3 * 100)),
        has_mpl=False,
    ),
    an=mrpd.CellSide(
        cl=an_cl, gdl=gdl_an,
        ch=mrpd.FlowChannel(height=1e-3, bulk_thermal_conductivity=100., width=1e-3, n_parallel=100, length=50e-4/(2e-3 * 100)),
        has_mpl=False,
    ),
    memb=mrpd.PFSAMembrane(
        thickness=25e-6,              # Nafion NR-211
        bulk_thermal_conductivity=0.3,
        ionomer=nafion,
    ),
)
print(cell.eps_ion, cell.eps_p)
```

## 2 — Model setup and initial conditions

```python
# Operating conditions — Table 5 of Vetter & Schumacher (2019)
T_OP = 343.15   # K  (70 °C)
P_OP = 1.5e5    # Pa (1.5 bar)
RH   = 0.9      # 90 % RH at both inlets
S_C  = 0.12     # cathode GDL/channel liquid saturation

# ── Inlet conditions — differential-cell approximation (S = 1000) ────────
STOICH = 100

# Conditions are defined first; model is instantiated with conditions= so that
# get_inputs() returns a CellSnapshot at each ODE step, enabling channel dynamics.
def _i_ref(t):
    i = model.current_density(t) if callable(model.current_density) else float(model.current_density)
    return max(float(i), 1.)

ca_cond = mrpd.InletAirConditions(
    temperature          = T_OP,
    backpressure         = P_OP,
    rh_ref_pressure      = P_OP,
    o2_molar_flow_rate   = lambda t: STOICH * np.maximum(_i_ref(t), 5000) * cell.area / (4 * ct.faraday),
    o2_dry_mole_fraction = 0.21,
    inlet_rh             = RH,
)
an_cond = mrpd.InletHydrogenConditions(
    temperature          = T_OP,
    backpressure         = P_OP,
    rh_ref_pressure      = P_OP,
    h2_molar_flow_rate   = lambda t: 2 * STOICH * np.maximum(_i_ref(t), 5000) * cell.area / (4 * ct.faraday),
    inlet_rh             = RH,
)
conditions = mrpd.CellConditions(current_density=0., ca=ca_cond, an=an_cond)

base = mrpd.CellBaseModel(
    # Pass conditions= to enable channel gas dynamics in rates_of_change
    transient_transport_model=mrpd.TransientCellModel(
        cell=cell, current_density=0., conditions=conditions,
    ),
    memb_model=mrpd.PFSAModel(),
    cl_model=mrpd.PtCCatalystLayerModel(),
    gas_diffusion_model=mrpd.PorousGasResistanceModel(),
    darcy_transport_model=mrpd.DarcyTransportModel(),
    voltage_model=mrpd.VoltageModel(),
)
model = base.transient_transport_model

y0 = model.initial_state_from_conditions(conditions.at(0.), s_ca=S_C)
```

## 3 — Steady-state polarization curve sweep

For each current density step `solve_steady_state` finds `x` such that
`rates_of_change(t, x) = 0` using a nonlinear root finder (MINPACK HYBRD by default).
Each step is warm-started from the previous solution.  The sweep stops when the
solver fails to converge or the cell voltage drops to zero.

The returned `sol.y` (shape `(n_states, 1)`) is directly compatible with
`base.postprocess`, so no time-averaging or index arithmetic is needed.


```python
CURRENT_DENSITIES = np.linspace(10, 20000, 101)   # A/m2  (0 to 2 A/cm2)

V_arr   = np.full(len(CURRENT_DENSITIES), np.nan)
y_store = np.full((len(CURRENT_DENSITIES), len(y0)), np.nan)

y = y0.copy()

for k, i_k in enumerate(CURRENT_DENSITIES):
    # Fix current density for this operating point (affects inlet flows via _i_ref)
    
    model.current_density      = float(i_k)
    conditions.current_density = float(i_k)
    
    sol = base.solve_steady_state(y, t=0.)
    if not sol.success:
        print(f"Step {k} (i={i_k/1e4:.3f} A/cm\u00b2): solver did not converge — stopping.")
        break

    st = base.postprocess(sol.y, i_density=float(i_k))
    V_k = float(st.V_cell[0])
    if V_k <= 0.:
        print(f"Step {k} (i={i_k/1e4:.3f} A/cm\u00b2): V_cell = {V_k:.3f} V ≤ 0 — stopping.")
        break

    V_arr[k]   = V_k
    y_store[k] = sol.y[:, 0]
    y          = sol.y[:, 0]   # warm-start next step

n_solved = int(np.sum(~np.isnan(V_arr)))
print(f"Solved {n_solved} / {len(CURRENT_DENSITIES)} operating points.")
```

## 4 — Polarization and power-density curves

```python
i_Acm2 = CURRENT_DENSITIES / 1e4
P_arr  = i_Acm2 * V_arr   # W/cm2

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 3.5))
fig.suptitle("PEMFC base case — Vetter & Schumacher (2019) conditions", fontsize=12)

ax1.plot(i_Acm2, V_arr, 'C0-o', label='$V_{cell}$')
ax1.set_xlabel('Current density (A/cm$^2$)')
ax1.set_ylabel('Cell voltage (V)')
ax1.set_ylim(0, 1.1)
ax1.axhline(0.6, color='grey', lw=0.8, ls='--', label='U = 0.6 V')
ax1.axvline(1.0, color='grey', lw=0.8, ls=':')
ax1.grid(True, alpha=0.3)
ax1.legend()

ax2.plot(i_Acm2, P_arr * 1e3, 'C1-o', label='Power density')
ax2.set_xlabel('Current density (A/cm$^2$)')
ax2.set_ylabel('Power density (mW/cm$^2$)')
ax2.grid(True, alpha=0.3)
ax2.set_ylim(0,1000)
ax2.legend()

fig.tight_layout()
plt.show()
```

## 5 — Comparison with Table 6 of Vetter & Schumacher (2019)

Derived quantities follow the definitions in the paper (Eqs. 29–31).  The operating
point at **U = 0.6 V** is obtained by linear interpolation of the polarization curve.

```python
# ── Interpolation helpers ─────────────────────────────────────────────────
valid  = ~np.isnan(V_arr)
i_v    = (CURRENT_DENSITIES / 1e4)[valid]
V_v    = V_arr[valid]

V_to_I = interp1d(V_v[::-1], i_v[::-1], kind='linear',
                   bounds_error=False, fill_value='extrapolate')
I_to_V = interp1d(i_v, V_v, kind='linear',
                   bounds_error=False, fill_value='extrapolate')

i_at_06  = float(V_to_I(0.6))
V_at_1   = float(I_to_V(1.0))

# Step whose steady-state voltage is closest to 0.6 V
idx_06 = int(np.nanargmin(np.abs(V_arr - 0.6)))
model.current_density = float(CURRENT_DENSITIES[idx_06])
conditions.current_density = float(CURRENT_DENSITIES[idx_06])
st06 = base.postprocess(y_store[idx_06, :, np.newaxis],
                         i_density=float(CURRENT_DENSITIES[idx_06]))

# ── Peak power and limit current density ─────────────────────────────────
i_Acm2     = CURRENT_DENSITIES / 1e4
peak_power = float(np.nanmax(i_Acm2 * V_arr))
i_lim      = float(i_v[-1]) if len(i_v) >= 1 else float('nan')

# ── State quantities at the 0.6 V operating point ─────────────────────────
c = cell

mea_layers = [c.an.gdl, c.an.cl, c.memb, c.ca.cl, c.ca.gdl]
mea_ix     = [l.ix for l in mea_layers]
mea_L      = np.array([l.thickness for l in mea_layers])
T_C        = st06.T[:, 0] - 273.15
T_peak     = float(np.max(T_C[mea_ix]))
T_avg      = float(np.dot(T_C[mea_ix], mea_L) / mea_L.sum())

ccm_ix   = [c.an.cl.ix, c.memb.ix, c.ca.cl.ix]
ccm_eps  = c.eps_ion[ccm_ix, 0]
ccm_L    = c.thickness[ccm_ix, 0]
lmbd06   = st06.lmbd[:, 0]
lmbd_min = float(np.min(lmbd06[ccm_ix]))
lmbd_avg = float(np.dot(ccm_eps * lmbd06[ccm_ix], ccm_L) / np.dot(ccm_eps, ccm_L))

J_PEM = float(st06.J[c.memb.ix, model.i_lmbd, 0]) * 1e5

def _sigma(ix):
    return base.memb_model.charge_conductivity(
        float(st06.f_v[ix, 0]), float(st06.T[ix, 0]), 'proton', c.memb)
sigma_left   = _sigma(c.an.cl.ix)
sigma_center = _sigma(c.memb.ix)
sigma_right  = _sigma(c.ca.cl.ix)
R_PEM = (c.memb.thickness / 6
         * (1/sigma_left + 4/sigma_center + 1/sigma_right)
         * 1e4 * 1e3)

rows = [
    ('Peak power density          (W/cm2)',    peak_power,  0.901),
    ('Limit current density        (A/cm2)',   i_lim,       1.960),
    ('Cell voltage at I=1 A/cm2       (V)',    V_at_1,      0.720),
    ('Current density at U=0.6 V  (A/cm2)',   i_at_06,     1.499),
    ('Peak temperature at 0.6 V      (C)',     T_peak,      70.90),
    ('Average temperature at 0.6 V   (C)',     T_avg,       70.36),
    ('Min. water content at 0.6 V    (-)',      lmbd_min,    3.72 ),
    ('Avg. water content at 0.6 V    (-)',      lmbd_avg,    6.68 ),
    ('Water flux through PEM (umol/cm2s)',      J_PEM,       3.05 ),
    ('Membrane resistance         (mOhm.cm2)', R_PEM,      83.9  ),
]

hdr = f"{'Quantity':<44} {'This model':>12} {'Vetter 2019':>12}"
print(hdr)
print('-' * len(hdr))
for name, ours, ref in rows:
    print(f"{name:<44} {ours:>12.3f} {ref:>12.3f}")
```

## 6 — Internal state vs. current density

Key quantities extracted from the quasi-steady state at the end of each
galvanostatic step, plotted against current density.

```python
c = cell
valid_k = np.where(~np.isnan(V_arr))[0]
i_v2    = (CURRENT_DENSITIES / 1e4)[valid_k]

# Postprocess all solved steady-states at once
Y_valid = y_store[valid_k].T                       # (n_states, n_valid)
I_valid = CURRENT_DENSITIES[valid_k]

# Set model to last solved current (needed for postprocess internal calls)
model.current_density      = float(I_valid[-1]) if len(I_valid) else 0.
conditions.current_density = model.current_density
state_v = base.postprocess(Y_valid, i_density=I_valid)

T_an  = state_v.T[c.an.cl.ix, :] - 273.15
T_mb  = state_v.T[c.memb.ix,  :] - 273.15
T_ca  = state_v.T[c.ca.cl.ix, :] - 273.15

lmbd_an = state_v.lmbd[c.an.cl.ix, :]
lmbd_mb = state_v.lmbd[c.memb.ix,  :]
lmbd_ca = state_v.lmbd[c.ca.cl.ix, :]

s_ca_cl  = state_v.s[c.ca.cl.ix,  :]
s_ca_gdl = state_v.s[c.ca.gdl.ix, :]

J_PEM_arr = state_v.J[c.memb.ix, model.i_lmbd, :] * 1e5

c_o2_ch = state_v.cg_k[c.ca.ch.ix, 0, :]   # O2, kmol/m³
c_h2_ch = state_v.cg_k[c.an.ch.ix, 2, :]   # H2, kmol/m³

def _R_PEM_col(j):
    def _sig(ix):
        return base.memb_model.charge_conductivity(
            float(state_v.f_v[ix, j]), float(state_v.T[ix, j]), 'proton', c.memb)
    return (c.memb.thickness / 6
            * (1/_sig(c.an.cl.ix) + 4/_sig(c.memb.ix) + 1/_sig(c.ca.cl.ix))
            * 1e4 * 1e3)

R_PEM_arr    = np.array([_R_PEM_col(j) for j in range(len(valid_k))])
eta_memb_arr = state_v.eta_memb * 1e3
eta_act_arr  = state_v.eta_act  * 1e3

fig, axes = plt.subplots(3, 3, figsize=(13, 10))
fig.suptitle("Internal state vs. current density — steady-state", fontsize=12)
kw = dict(ms=4)

ax = axes[0, 0]
ax.plot(i_v2, T_an, 'C2-o', **kw, label='An CL')
ax.plot(i_v2, T_mb, 'C1-o', **kw, label='Membrane')
ax.plot(i_v2, T_ca, 'C0-o', **kw, label='Ca CL')
ax.set_ylabel('Temperature (°C)'); ax.legend(fontsize=8); ax.grid(True, alpha=0.3)

ax = axes[0, 1]
ax.plot(i_v2, lmbd_an, 'C2-o', **kw, label='An CL')
ax.plot(i_v2, lmbd_mb, 'C1-o', **kw, label='Membrane')
ax.plot(i_v2, lmbd_ca, 'C0-o', **kw, label='Ca CL')
ax.axhline(22, color='grey', lw=0.8, ls='--', label=r'$\lambda_{liq}$')
ax.set_ylabel(r'Water content $\lambda$ (−)'); ax.legend(fontsize=8); ax.grid(True, alpha=0.3)

ax = axes[0, 2]
ax.plot(i_v2, s_ca_cl,  'C0-o',  **kw, label='Ca CL')
ax.plot(i_v2, s_ca_gdl, 'C0--s', **kw, label='Ca GDL')
ax.set_ylabel('Liquid saturation $s$ (−)'); ax.legend(fontsize=8); ax.grid(True, alpha=0.3)

ax = axes[1, 0]
ax.plot(i_v2, J_PEM_arr, 'C1-o', **kw)
ax.axhline(0, color='grey', lw=0.8, ls='--')
ax.set_ylabel(r'PEM water flux ($\mu$mol cm$^{-2}$ s$^{-1}$)'); ax.grid(True, alpha=0.3)

ax = axes[1, 1]
ax.plot(i_v2, R_PEM_arr, 'C3-o', **kw)
ax.set_ylabel(r'$R_{PEM}$ (m$\Omega\cdot$cm²)'); ax.grid(True, alpha=0.3)

ax = axes[1, 2]
ax.plot(i_v2, eta_memb_arr, 'C3-o', **kw, label=r'$\eta_{memb}$')
ax.plot(i_v2, eta_act_arr,  'C0-o', **kw, label=r'$\eta_{act}$')
ax.set_ylabel('Overpotential (mV)'); ax.legend(fontsize=8); ax.grid(True, alpha=0.3)

ax = axes[2, 0]
ax.plot(i_v2, c_o2_ch * 1e3, 'C0-o', **kw)
ax.set_ylabel(r'Ca channel $c_{O_2}$ (mol/m³)'); ax.grid(True, alpha=0.3)

ax = axes[2, 1]
ax.plot(i_v2, c_h2_ch * 1e3, 'C2-o', **kw)
ax.set_ylabel(r'An channel $c_{H_2}$ (mol/m³)'); ax.grid(True, alpha=0.3)

axes[2, 2].axis('off')
for ax in axes.flat:
    ax.set_xlabel('Current density (A/cm²)')

fig.tight_layout()
plt.show()
```

## 7 — Capillary pressure–saturation: Vetter Eq. 27 vs. Van Genuchten model

Vetter & Schumacher (2019) use an empirical two-exponential fit for Toray
TGP-H-060 (Eq. 27, from Nguyen et al. 2008 [62]), while our model uses the
Van Genuchten (VGM) expression.

**Sign conventions differ:**  
- Vetter: p_c = p_liquid − p_gas → negative in the hydrophobic regime  
- VGM (our model): capillary pressure as magnitude, always ≥ 0  

The liquid transport is driven by ∂p_c/∂s (right panel), which is comparable
between the two models in the operating range, though Vetter's fit has a
steeper gradient at low saturations.

## 8 — Solver benchmark: steady-state vs. IVP per operating point

Both methods are timed step-by-step with warm-starting from the previous solution,
starting from the same `y0`.

* **Steady-state** is benchmarked over all `n_solved` converged steps.
* **IVP** integrates `T_STEP = 500 s` per step (same as the channel notebook)
  but is capped at `N_BENCH_IVP` steps because the full sweep would take
  O(hours).  Mean time is extrapolated for the full sweep.


```python
import time

N_BENCH_IVP  = 20      # IVP steps to actually run (extrapolate the rest)
T_STEP_BENCH = 500.    # IVP integration window per step, same as channel notebook

bench_k     = np.where(~np.isnan(V_arr))[0]          # SS-converged step indices
bench_k_ivp = bench_k[:N_BENCH_IVP]                  # capped to available steps
n_ivp       = len(bench_k_ivp)                        # actual number of IVP steps run
i_bench     = (CURRENT_DENSITIES / 1e4)[bench_k]
n_full      = len(bench_k)

# ── Steady-state: time every converged step ───────────────────────────────
t_ss  = np.full(n_full, np.nan)
V_ss  = np.full(n_full, np.nan)   # voltage at each SS step
y = y0.copy()
for j, k in enumerate(bench_k):
    model.current_density      = float(CURRENT_DENSITIES[k])
    conditions.current_density = float(CURRENT_DENSITIES[k])
    t0 = time.perf_counter()
    sol = base.solve_steady_state(y, t=0.)
    t_ss[j] = time.perf_counter() - t0
    if sol.success:
        st = base.postprocess(sol.y, i_density=float(CURRENT_DENSITIES[k]))
        V_ss[j] = float(st.V_cell[0])
        y = sol.y[:, 0]

# ── IVP: time first n_ivp steps ──────────────────────────────────────────
t_ivp = np.full(n_ivp, np.nan)
V_ivp = np.full(n_ivp, np.nan)    # voltage at end of each IVP step
y = y0.copy()
for j, k in enumerate(bench_k_ivp):
    model.current_density      = float(CURRENT_DENSITIES[k])
    conditions.current_density = float(CURRENT_DENSITIES[k])
    t0 = time.perf_counter()
    sol_ivp = base.solve(y, t_span=(0., T_STEP_BENCH), max_step=100.)
    t_ivp[j] = time.perf_counter() - t0
    y = sol_ivp.y[:, -1]
    st_ivp = base.postprocess(sol_ivp.y[:, -1:], i_density=float(CURRENT_DENSITIES[k]))
    V_ivp[j] = float(st_ivp.V_cell[0])
    print(f"  IVP step {j+1:2d}/{n_ivp}  i={CURRENT_DENSITIES[k]/1e4:.2f} A/cm²  "
          f"t={t_ivp[j]*1e3:.0f} ms", end='\r')
print()

# ── Summary table ─────────────────────────────────────────────────────────
ss_mean  = float(np.nanmean(t_ss))
ss_total = float(np.nansum(t_ss))
ivp_mean = float(np.nanmean(t_ivp))
speedup  = ivp_mean / ss_mean

print(f"{'Method':<36} {'Mean (ms)':>10} {'Total (s)':>11}")
print('-' * 59)
print(f"{'Steady-state  (all steps)':<36} {ss_mean*1e3:>10.1f} {ss_total:>11.2f}")
print(f"{f'IVP (500 s), {n_ivp} steps measured':<36} {ivp_mean*1e3:>10.0f} "
      f"{float(np.nansum(t_ivp)):>11.2f}")
print(f"{f'IVP extrapolated ({n_full} steps)':<36} {'':>10} {ivp_mean*n_full:>11.1f}")
print(f"\nSpeedup (IVP mean / SS mean): {speedup:.1f}×")

# ── Figure 1: timing ──────────────────────────────────────────────────────
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(11, 4))
fig.suptitle(f"Solver wall-clock time per operating point  (IVP: T_step={T_STEP_BENCH:.0f} s)", fontsize=11)

ax1.plot(i_bench,         t_ss * 1e3,  'C0-o', ms=4, label=f'Steady-state ({n_full} pts)')
ax1.plot(i_bench[:n_ivp], t_ivp * 1e3, 'C1-s', ms=5, label=f'IVP ({n_ivp} pts)')
ax1.set_xlabel('Current density (A/cm²)')
ax1.set_ylabel('Wall-clock time (ms)')
ax1.set_title('Per-step timing')
ax1.legend(fontsize=9)
ax1.grid(True, alpha=0.3)

ivp_cum_extrap = np.arange(1, n_full + 1) * ivp_mean
ax2.plot(i_bench,         np.cumsum(t_ss),  'C0-o', ms=4, label='Steady-state')
ax2.plot(i_bench,         ivp_cum_extrap,   'C1--', lw=1.5,
         label=f'IVP extrapolated (mean={ivp_mean:.1f} s/pt)')
ax2.plot(i_bench[:n_ivp], np.cumsum(t_ivp), 'C1-s', ms=5, label=f'IVP measured ({n_ivp} pts)')
ax2.set_xlabel('Current density (A/cm²)')
ax2.set_ylabel('Cumulative wall-clock time (s)')
ax2.set_title('Cumulative timing')
ax2.legend(fontsize=9)
ax2.grid(True, alpha=0.3)

fig.tight_layout()
plt.show()

# ── Figure 2: polarization curve comparison (over measured IVP range) ─────
i_common = i_bench[:n_ivp]                # current densities where both are available
dV       = V_ss[:n_ivp] - V_ivp          # SS minus IVP

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(11, 4))
fig.suptitle(f"Polarization curve: steady-state vs. IVP  (first {n_ivp} pts)", fontsize=11)

ax1.plot(i_bench,    V_ss,       'C0-o', ms=4, lw=1.5, label='Steady-state (all pts)')
ax1.plot(i_common,   V_ivp,      'C1-s', ms=5, lw=1.5, label=f'IVP ({n_ivp} pts)')
ax1.set_xlabel('Current density (A/cm²)')
ax1.set_ylabel('Cell voltage (V)')
ax1.set_ylim(bottom=0)
ax1.axhline(0.6, color='grey', lw=0.8, ls='--')
ax1.legend(fontsize=9)
ax1.grid(True, alpha=0.3)

ax2.axhline(0, color='grey', lw=0.8, ls='--')
ax2.plot(i_common, dV * 1e3, 'C2-o', ms=4, lw=1.5)
ax2.set_xlabel('Current density (A/cm²)')
ax2.set_ylabel('V_SS − V_IVP  (mV)')
ax2.set_title('Voltage difference (SS − IVP)')
ax2.grid(True, alpha=0.3)

fig.tight_layout()
plt.show()
```
